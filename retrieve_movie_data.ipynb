### Access the New York Times API

```python
### Set the base URL
url = "https://api.nytimes.com/svc/search/v2/articlesearch.json?"

# Site example
# https://api.nytimes.com/svc/search/v2/articlesearch.json?q=election&api-key=yourkey

### Filter for movie reviews with "love" in the headline
### section_name should be "Movies"
### type_of_material should be "Review"
filter_query = 'section_name:"Movies" AND type_of_material:"Review" AND headline:"love"'

### Use a sort filter, sort by newest
sort = "newest"

### Select the following fields to return:
### headline, web_url, snippet, source, keywords, pub_date, byline, word_count
field_list = "headline,web_url,snippet,source,keywords,pub_date,byline,word_count"

### Search for reviews published between a begin and end date

begin_date = "20130101"
end_date = "20230531"


### Build URL

seach_url = url +"api-key=" + nyt_api_key + "&fq=" + filter_query + "&sort=" + sort + "&fl=" + field_list + "&begin_date=" + begin_date + "&end_date=" + end_date
# query_url is correctly constructed ---------------------------------------------------------------------------------------------------------------------------------> (2 points).
type(seach_url)
```

```python
# Checked page 0
# Checked page 1
# Checked page 2
# Checked page 3
# Checked page 4
# Checked page 5
# Checked page 6
# Checked page 7
# Checked page 8
# Checked page 9
# Checked page 10
# Checked page 11
# Checked page 12
# Checked page 13
# Checked page 14
# Checked page 15
# Checked page 16
# Checked page 17
# Checked page 18
# Checked page 19

### Create an empty list to store the reviews
review_list = []
# An empty list reviews_list is created -----------------------------------------------------------------------------------------------------------------------> (1 point).

# print("Review list")
# print(review_list)
# print('-----------------------')

### loop through pages 0-19
for pages in range(0,20):
    # A for loop is created to loop through 20 times ----------------------------------------------------------------------------------------------------------> (3 points).


    ### create query with a page number
    loop_url = seach_url + "&page=" + str(pages)
    # The query_url is extended to include a page --------------------------------------------------------------------------------------------------------------> (1 point).

    ### API results show 10 articles at a time
    ###################################################
    ################################################### Yes it does
    ###################################################

        ### Make a "GET" request and retrieve the JSON
    reviews = requests.get(loop_url).json()
    # A GET request is made to retrieve results and the JSON data is stored in a variable called reviews -----------------------------------------------------> (4 points).
    
    ### Add a twelve second interval between queries to stay within API query limits
    time.sleep(12)
    # A 12-second interval is used between queries ---------------------------------------------------------------------------------------------------------> (2 points).

    
    ### Try and save the reviews to the reviews_list
    try:
    # A try-except clause is used ----------------------------------------------------------------------------------------------------------------------------> (2 points).

        ### loop through the reviews["response"]["docs"] and append each review to the list
        for articles in reviews["response"]["docs"]:
            # Inside the try clause, there is a loop to loop through the reviews["response"]["docs"] list -----------------------------------------------------> (3 points).
            review_list.append(articles)
            # The reviews results are correctly appended to reviews_list --------------------------------------------------------------------------------------> (2 points).

        ### Print the page number that had no results then break from the loop
    except:
        print(f"Check Page : {pages} - Error")
        pass
        # The except clause prints out the page number that had no results, then breaks from the loop -----------------------------------------------------------> (2 points).

### Print the page that was just retrieved
    print(f"Check Page : {pages}")
    # The query page number is printed -----------------------------------------------------------------------------------------------------------------> (1 point).
```

```python
### Extract 'name' and 'value' from items in "keywords" column
def extract_keywords(keyword_list):
    extracted_keywords = ""
    for item in keyword_list:
        # Extract 'name' and 'value'
        keyword = f"{item['name']}: {item['value']};" 
        # Append the keyword item to the extracted_keywords list
        extracted_keywords += keyword
        # print(extracted_keywords)
    return extracted_keywords



### Fix the "keywords" column by converting cells from a list to a string
df_nyt_title = df_nyt.copy()
df_nyt_title["keywords"] = df_nyt_title["keywords"].apply(extract_keywords)
# The "keywords" column is correctly converted to string data using the supplied extract_keywords function --------------------------------------> (3 points).

df_nyt_title
```
